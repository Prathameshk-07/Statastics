```python
import numpy as np
from scipy import stats

data=[25,20,23,24,26,29,36,45,50,27,28,53,25,28,23,24,53,26,45]

#mean
Mean= np.mean(data)
print(f"Mean:{Mean: .2f}")

#calculate Median
Median= np.median(data)
print(f"Median:{Median: .2f} ")

#Mode
mode=stats.mode(data, keepdims=True)
print(f"Mode: {mode.mode[0]}")
print(f"Frequency: {mode.count[0]}")
```
